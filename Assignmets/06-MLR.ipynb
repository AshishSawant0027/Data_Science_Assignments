```python
x= df.iloc[:,1:]
y = df.iloc[:,0]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state= 2 , train_size= 0.8)
```

```python
for_model = pd.concat([x_train, y_train], axis = 1)
for_model
```

```python
for_model.columns
```

```python
import statsmodels.formula.api as smf
model = smf.ols('Price~ Age_08_04 + KM + Fuel_Type + HP +Automatic + cc + Doors + Gears + Weight', data= for_model ).fit()
```

```python
model.summary()
```

##### pvalue for cc column and Doors column is greater than level of significance so we accept the null hypothesis. so these two columns have no significance influence on Price column together. let us check if these values have any significance.

```python
model1 = smf.ols("Price~cc", data = for_model).fit()
model1.pvalues
```

```python
model2 = smf.ols("Price~Doors", data = for_model).fit()
model2.pvalues
```

```python
# the pvalues of cc and Doors separately suggest a significant influence on dependent variable.
#this implies cc and doors together 'may' be  generating the problem of collinearity.
```

```python
model3 = smf.ols(" Price~ cc+ Doors", data = for_model).fit()
print(model3.pvalues)
```

```python
# from above we can see that both  cc and doors is not the problem. but the combination of them with some other rows  could be. so we remove cc and doors
#it was found that Automatic column had pvalue greater than 0.05 after removing cc and doors. so we removed it too,which did not hampen any change to accuracy
```

```python
model4 = smf.ols("Price~ Age_08_04 + KM + Fuel_Type + HP  + Gears + Weight", data= for_model ).fit() 
model4.summary()
```

```python
#coefficient values and intercept.
model4.params
```

```python
model4.rsquared
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
y_pred = model4.predict(x_test)
```

```python
mse= mean_squared_error(y_test, y_pred)
rmse =np.sqrt(mse)
rmse
```

### 1. **Normalization & Standardization**:
- **Normalization**: Scales data to a fixed range, usually [0, 1]. Useful when features have different units or scales. 
  - Formula:  
    #### Xnorm = (X - Xmin) / (Xmax - Xmin)
  
- **Standardization**: Centers data around a mean of 0 and standard deviation of 1. Helps in algorithms that assume normally distributed features. 
  - Formula:  
    #### Xstd = X-mu/sigma
Both improve model performance by making features comparable in scale.

---

### 2. **Addressing Multicollinearity**:
- **Remove correlated features**: Drop one of the correlated variables.
- **Principal Component Analysis (PCA)**: Converts correlated variables into uncorrelated components.
- **Variance Inflation Factor (VIF)**: Remove variables with high VIF values (>10).
- **Ridge Regression**: Uses L2 regularization to shrink coefficients.
- **Lasso Regression**: Uses L1 regularization for feature selection and reduces multicollinearity.
