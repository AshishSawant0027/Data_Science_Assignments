```python
from apyori import apriori
```

```python
df1.shape
```

```python
records = []
for i in range(0,7501):
    records.append([str(df1.values[i,j]) for j in range (0, 20)])
```

```python
records[0:2]
```

```python
#apply apriori algorithm
association_rules = apriori(records,min_support = 0.0045 , min_confidence = 0.2 , min_lift= 3)
association_results = list(association_rules)
```

```python
association_results[1]
```

items_base= antecedent, 

items_add = consequent items
mushroom cream sauce --> escalope
1. support = 0.0057 implies that approximately 0.573% of all transactions contain both escalope and mushroom cream sauce.
2. confidence = 0.30 implies that there is approximately a 30.07% chance that if a transaction contains mushroom cream sauce, it also contains escalope.
3.  lift - 3.79 implies that transactions containing mushroom cream sauce are approximately 3.79 times more likely to also contain escalope 

```python
association_results = sorted(association_results, key = lambda x: x[1], reverse = True)
```

```python
for item in association_results:
    pair = item.items
    items = [x for x in pair]
    # Process each OrderedStatistic object
    for stat in item.ordered_statistics:
        antecedent = list(stat.items_base)
        consequent = list(stat.items_add)
        
        rule = str(antecedent) + ' -> ' + str(consequent)
        
        print('Rule: ' + rule)
        print('Support: ' + str(item.support))
        print('Confidence: ' + str(stat.confidence))
        print('Lift: ' + str(stat.lift))
        print('=============================================')
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
df2
```

```python
df2.drop( 'None', axis =1 , inplace = True)
df2
```

```python
frequent_itemsets = apriori(df2, min_support=0.0045, use_colnames = True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, min_threshold= 0.4)
rules.sort_values('lift' , ascending = False)
```

#### What Are Some Limitations or Challenges of Association Rules Mining?
1. Scalability: Increased data size leads to exponential growth in the number of potential itemsets and rules, making computations costly.

2. Data Sparsity: Many itemsets may have low support in large datasets, resulting in sparse and less meaningful rule generation.

3. Rule Interpretation: The high number of rules can be overwhelming, requiring additional filtering to identify the most useful ones.

4. Parameter Sensitivity: Results depend heavily on parameters like min_support and min_confidence, affecting rule quality.

5. Overfitting: High-confidence rules may fit the training data too closely, reducing generalizability to new data.

6. Lack of Temporal Information: Association rules do not capture the order or timing of transactions.

7. Not Suitable for All Types of Data: Association rule mining is less effective for continuous or non-categorical data without preprocessing.

8. Context Dependency: Rules may be meaningful in one context but not in another, making their applicability context-specific.
